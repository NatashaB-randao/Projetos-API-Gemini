```python
# Imporata a biblioteca do Google Generative AI
import google.generativeai as genai
```

```python
# Função principal do assistente
def main():
    # Configurações do assistente
    assistente_falante = True  # Define se o assistente deve falar
    ligar_microfone = True      # Define se o microfone deve ser ativado

    # Configura a chave da API do Google Gemini
    genai.configure(api_key="AIzaSyAaCIFqxIiLuLPl001BuHTaKjVDgBV0low")

    # Lista todos os modelos disponíveis e imprime aqueles que suportam geração de conteúdo
    for m in genai.list_models():
        if 'generateContent' in m.supported_generation_methods:
            print(m.name)

    # Cria um modelo gerativo usando o modelo 'gemini-pro'
    model = genai.GenerativeModel('gemini-pro')
    # Inicia um chat com o modelo
    chat = model.start_chat(history=[])

    ### Configura a voz do assistente
    if assistente_falante:
        import pyttsx3  # Biblioteca para síntese de voz
        engine = pyttsx3.init()  # Inicializa o motor de síntese de voz

        # Obtém as vozes disponíveis e configura a velocidade
        voices = engine.getProperty('voices')
        engine.setProperty('rate', 180)  # Velocidade da fala

        # Imprime a lista de vozes disponíveis
        print("\nLista de Vozes - Verifique o número\n")
        for indice, vozes in enumerate(voices):  # Lista todas as vozes
            print(indice, vozes.name)

        voz = 1  # Seleciona a voz desejada (por índice)
        engine.setProperty('voice', voices[voz].id)  # Define a voz do assistente

    # Ativa o reconhecimento de fala se o microfone estiver ligado
    if ligar_microfone:
        import speech_recognition as sr  # Biblioteca para reconhecimento de fala
        r = sr.Recognizer()  # Cria um reconhecedor
        mic = sr.Microphone()  # Cria um objeto microfone

    # Mensagem de boas-vindas
    bem_vindo = "# Bem Vindo ao Assistente Mil Grau com Gemini AI #"
    print("")
    print(len(bem_vindo) * "#")  # Imprime uma linha de hashtags
    print(bem_vindo)  # Imprime a mensagem de boas-vindas
    print(len(bem_vindo) * "#")  # Imprime outra linha de hashtags
    print("###   Digite 'desligar' para encerrar    ###")  # Instrução ao usuário
    print("")

    # Início do loop principal do assistente
    while True:
        # Se o microfone estiver ativado, escuta o que o usuário diz
        if ligar_microfone:
            with mic as fonte:
                r.adjust_for_ambient_noise(fonte)  # Ajusta para ruído ambiente
                print("Fale alguma coisa (ou diga 'desligar')")
                audio = r.listen(fonte)  # Escuta o áudio
                print("Enviando para reconhecimento")
                try:
                    # Reconhece o áudio usando o Google
                    texto = r.recognize_google(audio, language="pt-BR")
                    print("Você disse: {}".format(texto))  # Imprime o texto reconhecido
                except Exception as e:
                    print("Não entendi o que você disse. Erro", e)  # Mensagem de erro
                    texto = ""
        else:
            # Caso o microfone não esteja ligado, solicita entrada de texto
            texto = input("Escreva sua mensagem (ou #sair): ")

        # Verifica se o usuário deseja desligar o assistente
        if texto.lower() == "desligar":
            break

        # Envia a mensagem para o modelo e obtém a resposta
        response = chat.send_message(texto)
        print("Gemini:", response.text, "\n")  # Imprime a resposta do modelo

        # Se o assistente falar, reproduz a resposta
        if assistente_falante:
            engine.say(response.text)  # Fala a resposta
            engine.runAndWait()  # Espera a fala terminar

    # Mensagem ao encerrar o chat
    print("Encerrando Chat")

# Verifica se o script está sendo executado diretamente
if __name__ == '__main__':
    main()  # Chama a função principal

```
